```python
#Handling Outliers
def outlier_capping(x):
    x = x.clip(lower=x.quantile(0.01), upper=x.quantile(0.99))
    return x

finaldf2=finaldf1.apply(lambda x: outlier_capping(x))
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
#standardize the data

sc = StandardScaler()
sc = sc.fit(finaldf2)
finaldfstd = pd.DataFrame(sc.transform(finaldf1), columns = finaldf1.columns)
```

```python
finaldfstd
```

```python
km4 = KMeans(n_clusters=4).fit(finaldfstd)
```

```python
km_3 = KMeans(n_clusters=3, random_state=123).fit(finaldfstd)
km_4 = KMeans(n_clusters=4, random_state=123).fit(finaldfstd)
km_5 = KMeans(n_clusters=5, random_state=123).fit(finaldfstd)
km_6 = KMeans(n_clusters=6, random_state=123).fit(finaldfstd)
km_7 = KMeans(n_clusters=7, random_state=123).fit(finaldfstd)
km_8 = KMeans(n_clusters=8, random_state=123).fit(finaldfstd)
```

```python
finaldfstd['Cluster_3'] = km_3.labels_
finaldfstd['Cluster_4'] = km_4.labels_
finaldfstd['Cluster_5'] = km_5.labels_
finaldfstd['Cluster_6'] = km_6.labels_
finaldfstd['Cluster_7'] = km_7.labels_
finaldfstd['Cluster_8'] = km_8.labels_
```

```python
cluster_range = range( 1, 20)
cluster_errors = []

for num_clusters in cluster_range:
    clusters = KMeans( num_clusters )
    clusters.fit(finaldfstd)
    cluster_errors.append( clusters.inertia_ )
```

```python
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )

clusters_df[0:10]
```

```python
# allow plots to appear in the notebook
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```
