```python
basket = opdf.groupby(["order_id","product_category_name"])["order_item_id"].sum().unstack().reset_index().fillna(0).set_index("order_id")
```

```python
basket
```

```python
#we need to consolidate the items into 1 transaction per row with each product 1 hot encoded.
# Convert the units to 1 hot encoded values
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1    
```

```python
basket_sets = basket.applymap(encode_units)
```

```python
basket_sets.head()
```

```python
# Build up the frequent items
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.001, use_colnames=True)
```

```python
frequent_itemsets
```

### we can see that frequent item sets doesnt have any combinations of 2 or 3 even when the min support is 1 percent  

```python
#Generate the rules with their corresponding support, confidence and lift:
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules
```
